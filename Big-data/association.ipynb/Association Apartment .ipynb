### Select & filter columns & Clean Data
- Column : ID , Year (2020 - 2024), Timestamp, Primary Type, Community Area, Arrest, Beat
- Year between 2020 - 2024 (5 Years)

**à¹€à¸¥à¸·à¸­à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆ Arrest = True**

```python
from pyspark.sql.functions import col

# à¹€à¸¥à¸·à¸­à¸à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹à¸¥à¸°à¸à¸£à¸­à¸‡à¸•à¸²à¸¡ Timestamp + Arrest = true
df_selected = df_with_timestamp.select(
    "ID", "Year", "Timestamp", "Primary Type", "Location Description",
    "Community Area", "Arrest", "Beat"
).filter(
    (col("Timestamp") >= "2013-01-01 00:00:00") &
    (col("Timestamp") <= "2024-12-31 23:59:59") &
    (col("Arrest") == True)
)

# à¹à¸ªà¸”à¸‡à¸•à¸±à¸§à¸­à¸¢à¹ˆà¸²à¸‡ 5 à¹à¸–à¸§
df_selected.show(5, truncate=False)
```

**à¹€à¸¥à¸·à¸­à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸¡à¸² 5 à¸›à¸µ**

```python
df_FiveYear = df_selected.filter(
    (col('Timestamp') >= '2020-01-01 00:00:00') &
    (col('Timestamp') <= '2024-12-31 23:59:59')
)
```

**à¸™à¸±à¸šà¸ˆà¸³à¸™à¸§à¸™à¹à¸–à¸§ 5 à¸›à¸µ**

```python
df_FiveYear.count()
```

**à¸à¸£à¸­à¸‡à¹à¸¥à¸°à¹€à¸•à¸£à¸µà¸¢à¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ Apartment à¸—à¸³à¹€à¸›à¹‡à¸™Basketà¹€à¸žà¸·à¹ˆà¸­à¸™à¸³à¹„à¸›à¹ƒà¸Šà¹‰à¸§à¸´à¹€à¸„à¸£à¸²à¸°à¸«à¹Œ Association**

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# 1. à¹€à¸¥à¸·à¸­à¸à¹€à¸‰à¸žà¸²à¸°à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆà¹€à¸à¸´à¸”à¹ƒà¸™ Apartment
df_apartment = df_FiveYear.filter(F.col("Location Description") == "APARTMENT")

# 2. à¸ªà¸£à¹‰à¸²à¸‡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ 'Year' à¸ˆà¸²à¸à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸§à¸±à¸™à¸—à¸µà¹ˆ (à¸–à¹‰à¸²à¸¢à¸±à¸‡à¹„à¸¡à¹ˆà¸¡à¸µ)
# à¸ªà¸¡à¸¡à¸•à¸´à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸Šà¸·à¹ˆà¸­ 'Date' à¸«à¸£à¸·à¸­ 'Date Occurred' â€” à¸›à¸£à¸±à¸šà¸•à¸²à¸¡à¸ˆà¸£à¸´à¸‡à¸™à¸°
df_apartment = df_apartment.withColumn("Year", F.year(F.col("Timestamp")))

# 3. à¸à¸¥à¸¸à¹ˆà¸¡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸•à¸²à¸¡à¸›à¸µ à¹à¸¥à¹‰à¸§à¸£à¸§à¸¡ Primary Type à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”à¹ƒà¸™à¸›à¸µà¸™à¸±à¹‰à¸™à¹€à¸›à¹‡à¸™ basket
df_basket_apartment = (
    df_apartment.groupBy("Year")
    .agg(F.collect_set("Primary Type").alias("basket"))
    .orderBy("Year")
)

# 4. à¹€à¸žà¸´à¹ˆà¸¡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ NO (à¹€à¸£à¸µà¸¢à¸‡à¸¥à¸³à¸”à¸±à¸šà¸›à¸µ)
windowSpec = Window.orderBy("Year")
df_basket_apartment = df_basket_apartment.withColumn("NO", F.row_number().over(windowSpec))

# 5. à¹€à¸£à¸µà¸¢à¸‡à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¹ƒà¸«à¹‰à¸ªà¸§à¸¢
df_basket_apartment = df_basket_apartment.select("NO", "Year", "basket")

df_basket_apartment.show(truncate=False)
```

### Association Process

```python
from pyspark.sql import functions as F

df_basket_apartment.select(F.size("basket").alias("len")).agg(F.max("len")).show()
print("à¸ˆà¸²à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸—à¸µà¹ˆà¸žà¸£à¹‰à¸­à¸¡à¸—à¸³ Association à¸¡à¸µà¸­à¸¢à¸¹à¹ˆ:", df_basket_apartment.count(), "à¹à¸–à¸§")
```

**à¸—à¸³à¹‚à¸¡à¹€à¸”à¸¥ Association**

```python
from pyspark.ml.fpm import FPGrowth
from pyspark.sql.functions import col

# --- à¸£à¸±à¸™ FP-Growth ---
fpGrowth = FPGrowth(itemsCol="basket", minSupport=0.75, minConfidence=1)
model = fpGrowth.fit(df_basket_apartment)
```

**--- à¸™à¸±à¸šà¸ˆà¸³à¸™à¸§à¸™ Association Rules ---**

```python

num_rules = model.associationRules.count()
print(f"à¸ˆà¸³à¸™à¸§à¸™ Association Rules: {num_rules}")
```

### à¸”à¸¹à¸à¸Žà¸—à¸µà¹ˆ Confidence = 1, Lift > 1, Itemset <= 3 

```python
from pyspark.sql.functions import col, size, concat_ws

# à¸à¸£à¸­à¸‡à¸à¸Ž: confidence = 1, lift > 1, antecedent â‰¤ 3 items
rules_filtered = model.associationRules.filter(
    (col("confidence") == 1.0) &
    (col("lift") > 1.0) &
    (size(col("antecedent")) <= 3)
)

# à¸£à¸§à¸¡ antecedent/consequent à¹€à¸›à¹‡à¸™ string à¸­à¹ˆà¸²à¸™à¸‡à¹ˆà¸²à¸¢
rules_filtered_str = rules_filtered.withColumn("antecedent", concat_ws(", ", col("antecedent"))) \
                                   .withColumn("consequent", concat_ws(", ", col("consequent")))

# à¸ªà¸±à¹ˆà¸‡ collect à¹à¸¥à¸° sort à¸•à¸²à¸¡ lift-desc
all_rules = rules_filtered_str.orderBy(col("lift").desc(), col("confidence").desc()).collect()

# à¹à¸ªà¸”à¸‡à¸œà¸¥
for row in all_rules:
    print(f"{row['antecedent']} => {row['consequent']} | confidence: {row['confidence']:.2f} | lift: {row['lift']:.2f}")
```

**Export à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸›à¹‡à¸™ Excel**

```python
# à¸•à¸´à¸”à¸•à¸±à¹‰à¸‡à¹„à¸¥à¸šà¸£à¸²à¸£à¸µà¸—à¸µà¹ˆà¹ƒà¸Šà¹‰à¹€à¸‚à¸µà¸¢à¸™ Excel (openpyxl) à¹à¸¥à¹‰à¸§à¹à¸›à¸¥à¸‡ all_rules à¹ƒà¸«à¹‰à¹€à¸›à¹‡à¸™ Pandas DataFrameà¹à¸¥à¹‰à¸§ export à¹€à¸›à¹‡à¸™ Excel à¹„à¸”à¹‰
import pandas as pd
rules_list = [(row['antecedent'], row['consequent'], row['confidence'], row['lift']) for row in all_rules]
rules_df = pd.DataFrame(rules_list, columns=['Antecedent', 'Consequent', 'Confidence', 'Lift'])
rules_df.to_excel("association_rules_apartment.xlsx", index=False)
```
